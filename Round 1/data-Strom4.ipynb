```python
grouped_df1 = df.groupby(['item_description', 'item_price']).agg({'item_description': 'count'})

# rename the column and reset the index
grouped_df1 = grouped_df1.rename(columns={'item_description': 'Count'}).reset_index()

# print the new DataFrame
print(grouped_df1.head())
```

```python
# calculate average invoice amount for each shop
avg_invoice_amount = df.groupby('shop_id')['item_price'].mean()
avg_invoice_amount=pd.DataFrame(avg_invoice_amount)

#avg_invoice_amount.set_index('shop_id', inplace=True)
avg_invoice_amount.rename(columns={'item_price': 'avg_invoice_amount'}, inplace=True)

# calculate frequency of transactions for each shop
freq_transactions = df.groupby('shop_id')['invoice_id'].nunique()
freq_transactions=pd.DataFrame(freq_transactions)

#freq_transactions.set_index('shop_id', inplace=True)
freq_transactions.rename(columns={'invoice_id': 'freq_transactions'}, inplace=True)

print(avg_invoice_amount)
print(freq_transactions)
```

```python
# df = df.sort_values(['shop_id','item_description','customer_id'])
# df1 = df1.sort_values('shop_id')
# Convert the transaction date column to datetime format
df['transaction_date'] = pd.to_datetime(df['transaction_date'])

# Group the data by shop and month, and calculate the monthly transaction totals per shop
monthly_totals = df.groupby(['shop_id', pd.Grouper(key='transaction_date', freq='M')])['quantity_sold'].sum()

# Calculate the 3-month moving average of monthly transaction totals per shop
monthly_moving_avg = monthly_totals.groupby('shop_id').rolling(window=3).mean()

# Group the data by shop and day, and calculate the daily transaction totals per shop
daily_totals = df.groupby(['shop_id', 'transaction_date'])['quantity_sold'].sum()

# Calculate the 7-day moving average of daily transaction totals per shop
daily_moving_avg = daily_totals.groupby('shop_id').rolling(window=7).mean()

# Print the monthly and daily transaction totals and moving averages per shop
print('Monthly transaction totals per shop:')
print(monthly_totals)
print('\n3-month moving average of monthly transaction totals per shop:')
print(monthly_moving_avg)
print('\nDaily transaction totals per shop:')
print(daily_totals)
print('\n7-day moving average of daily transaction totals per shop:')
print(daily_moving_avg)
```

```python
# Drop the missing rows from the DataFrame
monthly_moving_avg = monthly_moving_avg.dropna()

# Print the cleaned DataFrame
print(monthly_moving_avg)

print(monthly_moving_avg.shape)

# Drop the missing rows from the DataFrame
daily_moving_avg = daily_moving_avg.dropna()

# Print the cleaned DataFrame
print(daily_moving_avg)

print(daily_moving_avg.shape)
```

```python
# Group the data by the first level index (shop id) and calculate the mean value for each group
daily_shop_averages = daily_moving_avg.groupby(level=0).mean()

# Print the resulting averages
print(daily_shop_averages)
print(monthly_moving_avg)
```

```python
# Count the number of repeat customers for each shop
repeat_customers_by_shop = df.groupby(['shop_id', 'customer_id'])['transaction_date'].count()
repeat_customers_by_shop = repeat_customers_by_shop[repeat_customers_by_shop > 1]
repeat_customers_by_shop = repeat_customers_by_shop.groupby('shop_id').count()

# Count the number of unique customers for each shop
unique_customers_by_shop = df.groupby('shop_id')['customer_id'].nunique()

# Calculate the average number of repeat customers for each shop
average_repeat_customers_by_shop = repeat_customers_by_shop / unique_customers_by_shop
average_repeat_customers_by_shop = average_repeat_customers_by_shop.reset_index()
#average_repeat_customers_by_shop = average_repeat_customers_by_shop.rename(columns={'customer_id': 'count'})
average_repeat_customers_by_shop = average_repeat_customers_by_shop.rename(columns={0:'average_repeat_customers_by_shop'})
# Print the result
print(average_repeat_customers_by_shop)
```
